```python
Facematrix = []
Facelabel = []
for key,val in faces.items():
    Facematrix.append(val.flatten())
    Facelabel.append(key.split("/")[0])
 
# Create facematrix as (n_samples,n_pixels) matrix
Facematrix = np.array(Facematrix)
Facematrix
```

```python
facematrix=[]
for i in facematrix1:
    i=i-ty 
    facematrix.append(i)
facematrix
```

```python
facelabel=[]
for i in range(len(Facematrix)):
    if(i>29):
        facelabel.append(Facelabel[i])
facelabel
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(matrix, facelabel, test_size = 0.25, random_state = 42)
```

```python
from sklearn.decomposition import PCA
 
pca = PCA().fit(matrix)
print(pca.explained_variance_ratio_)
```

```python
n_components = 150
eigenfaces = pca.components_[:n_components]
dy=eigenfaces[68]
print(len(dy))
pam=np.resize(dy,(1275,1650))
pam=np.array(pam)
plt.imshow(pam,cmap='gray')
plt.show
```

```python
fig1, axes1 = plt.subplots(4,4,sharex=True,sharey=True,figsize=(8,10))
for i in range(16):
    dy=eigenfaces[i*7]
    pam=np.resize(dy,(1275,1650))
    pam=np.array(pam)
    axes1[i%4][i//4].imshow(pam, cmap="gray")
plt.show()
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
y_pred = knn.score(X_test_pca, y_test) 
```

```python
y_pred = knn.predict(X_test_pca)
y_pred
```

```python
y_test
```

```python
dr1=y_pred
dr2=y_test
sum=0
li1=[]
li2=[]
for i in range(len(dr1)):
    x1=dr1[i]
    x2=dr2[i]
    li1.append(x1[0])
    li2.append(x2[0])
    if(x1[0]==x2[0]):
        sum+=1 
ty=sum/len(dr1)
print(sum)
print(len(dr1))
print("Predicted")
print(li1)
print("Testing")
print(li2)
print(ty*100)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import ConfusionMatrixDisplay
cm= confusion_matrix(li2, li1)
cm_display=ConfusionMatrixDisplay(cm)
cm_display.plot()
plt.show()
```

```python
lis=['C','D','F','G','H','I','J','K','L','M']
for i in lis:
    su=0
    sin=0
    for j in range(len(li1)):
        if(i==li2[j]):
            su+=1
            if(i==li1[j]):
                sin+=1 
    print("The Accuracy for {} :".format(i))
    print((sin/su)*100)
    
```

```python
from sklearn.neighbors import KNeighborsClassifier
accuracy=[]
for i in range(10):
    knn = KNeighborsClassifier(n_neighbors=i+1)
    knn.fit(X_train_pca, y_train)
    y_pred = knn.predict(X_test_pca)
    dr1=y_pred
    dr2=y_test
    sum=0
    li1=[]
    li2=[]
    for i in range(len(dr1)):
        x1=dr1[i]
        x2=dr2[i]
        li1.append(x1[0])
        li2.append(x2[0])
        if(x1[0]==x2[0]):
            sum+=1 
    ty=sum/len(dr1)
    y=ty*100 
    accuracy.append(y)
accuracy=np.array(accuracy)
print(accuracy)
rt=[j+1 for j in range(10)]
rt=np.array(rt)
plt.plot(rt,accuracy)
plt.xlabel("Neighbors")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Neighbors")
plt.grid()
plt.show()


```
